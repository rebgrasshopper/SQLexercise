```python
#3. Write an if statement that asks for the user's name via input() function. If the name is "Bond" make it print "Welcome on board 007." Otherwise make it print "Good morning NAME". (Replace Name with user's name) 
name = input("What's your name? ")
if name.lower() == 'bond':
    print("Welcome on board 007")
elif not name:
    print("Well, whoever you are, it's nice to see you.")
else:
    print(f"Good morning {name.upper()}")
```
